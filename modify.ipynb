```python
import base64
import pandas as pd
import os

# Function to convert an image to Base64
def image_to_base64(image_path):
    try:
        with open(image_path, 'rb') as image_file:
            encoded_image = base64.b64encode(image_file.read()).decode('utf-8')
        return encoded_image
    except FileNotFoundError:
        return None  # Return None if the image is not found

# Function to add a base64 image column to a DataFrame
def add_base64_image_column(df, image_folder, image_column_name, image_id_column):
    # Create a new column to store base64-encoded images
    df[image_column_name] = df[image_id_column].apply(
        lambda x: image_to_base64(os.path.join(image_folder, f"{x}.jpg"))
    )

    return df

# Sample usage
if __name__ == "__main__":

    df = pd.read_csv("final_data.csv")

    # Folder containing the images
    image_folder = 'C:/Users/abhi9/Desktop/poster sample'  # Replace with your image folder path

    # Add the Base64 encoded image column to the DataFrame
    df = add_base64_image_column(df, image_folder, 'poster_image_base64', 'tconst')
```
