```python
similarity_df=pd.read_csv("top_100_similar_movies.csv")
```

```python
def top_movies(movie):
    movie_index=movies[movies['originalTitle']==movie].index[0]
    l=similarity_df.iloc[movie_index]
    d=list(l)
    for i in d[1:10]:
        print(movies.iloc[i].originalTitle)
```
